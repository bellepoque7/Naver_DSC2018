## enumertate 란?  range 함수의 진화버젼

```python
subject = ['R', 'Python','통계','살려줘']
for i,v in enumerate(subject):
    print(i,v)
```

## Map & reduce

```python
# map(함수, 리스트)
a = map(lambda x:x**2, range(5))
squares = list(a)
```

```python
squares
```

```python
# reduce(함수, 순서형 자료 )
from functools import reduce 
reduce(lambda x,y:x+y,[0,1,2,3,4])
```

```python
reduce(lambda x, y: y + x, 'abcde')
```
